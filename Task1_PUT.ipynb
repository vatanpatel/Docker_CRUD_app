```python
import requests
import pandas as pd
import json
import numpy as np
```

```python
data = pd.read_csv("Greendeck SE Assignment Task 1.csv", encoding= 'unicode_escape')
```

```python
for i in np.arange(len(data)):
    response = requests.post("http://127.0.0.1:5000/data", json = {
        "brand_name": data['brand_name'][i],
        "classification_l1": data['classification_l1'][i],
        "classification_l2": data['classification_l2'][i],
        "classification_l3": data['classification_l3'][i],
        "classification_l4": data['classification_l4'][i],
        "currency": data['currency'][i],
        "id": int(i),
        "image_url": data['image_url'][i],
        "name": data['name'][i],
        "offer_price_value": data['offer_price_value'][i],
        "regular_price_value": data['regular_price_value'][i]
    })
```
